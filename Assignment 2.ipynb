## Load Data

```python
import pandas as pd

df = pd.read_csv("patients.csv")
df.head()
```

## Queries

### 1. Search by Diagnosis

```python
back_pain_patients = df[df["Diagnosis"] == "Back Pain"]
print(back_pain_patients)
```

### 2. Patients with Low Sessions (Missed >2 sessions logic)

```python
missed_patients = df[df["SessionsCompleted"] < 3]
print(missed_patients)
```

### 3. Upcoming Week Appointments

```python
df["NextAppointment"] = pd.to_datetime(df["NextAppointment"])

upcoming = df[
    (df["NextAppointment"] >= "2026-05-01") &
    (df["NextAppointment"] <= "2026-05-07")
]

print(upcoming)
```

## Querying and Data Retrieval

### Search by Diagnosis
This query filters patients based on a specific diagnosis such as Back Pain. It helps physiotherapists quickly identify patients with similar conditions.

### Missed Sessions
Patients with fewer completed sessions are identified. This helps in tracking patient compliance and follow-ups.

### Upcoming Appointments
Patients scheduled within a specific date range are extracted. This helps in planning weekly schedules efficiently.

## Visualization

### Pie Chart – Diagnosis Distribution

```python
import matplotlib.pyplot as plt

diagnosis_counts = df["Diagnosis"].value_counts()

plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct="%1.1f%%")
plt.title("Diagnosis Distribution")
plt.show()
```

### Bar Chart – Sessions Completed

```python
plt.bar(df["PatientID"], df["SessionsCompleted"])
plt.xlabel("Patient ID")
plt.ylabel("Sessions Completed")
plt.title("Sessions Completed by Patients")
plt.xticks(rotation=90)
plt.show()
```

## Visualization and Reporting

### Diagnosis Distribution (Pie Chart)
This chart shows the proportion of patients suffering from different conditions. It helps in understanding the most common health issues.

### Sessions Completed (Bar Chart)
This graph represents how many therapy sessions each patient has completed. It helps in monitoring patient progress.
